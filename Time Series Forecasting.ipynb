```python
# now we will split our data into train and test to do time series forecasting
train = df[df.index.year <= 2015]
test = df[df.index.year > 2015]
```

```python
print(train.shape)
print(test.shape)
```

```python
# plotting the price of testing and trainig data
train['Price'].plot(figsize=(13,5), fontsize=15)
test['Price'].plot(figsize=(13,5), fontsize=15)
plt.grid()
plt.legend(['Training Data', 'Testing Data'])
plt.show()
```

```python
# now we will train our model
# we will train two model Linear Regression and naive Bayes
# we will form our train and test data
train_time = [i+1 for i in range(len(train))]
test_time = [i+ len(train)+1 for i in range(len(test))]
len(train_time), len(test_time)
```

```python
LR_train = train.copy()
LR_test = test.copy()
```

```python
LR_train['time'] = train_time
LR_test['time'] = test_time
```

```python
lr = LinearRegression()
lr.fit(LR_train[['time']], LR_train['Price'].values)
```

```python
test_predict_model1 = lr.predict(LR_test[['time']])
LR_test['forecast'] = test_predict_model1

plt.figure(figsize=(14,6))
plt.plot(train['Price'], label = 'train')
plt.plot(test['Price'], label = 'test')
plt.plot(LR_test['forecast'], label = 'reg on time_test data')
plt.legend(loc = 'best')
plt.grid()
plt.show()
```

```python
# we will now find MAPE which is Measure of Prediction of Accuracy of a Forecasting method in a statistical model
def mape(actual, pred):
    return round((np.mean(abs(actual - pred)/ actual)) * 100, 2)
```

```python
mape_model1_test = mape(test['Price'].values, test_predict_model1)
print('MAPE is %3.3f'%(mape_model1_test),'%')
```

```python
result = pd.DataFrame({'Test Mape (%)' : [mape_model1_test]}, index=['RegressionOnTime'])
result
```

```python
naive_train = train.copy()
naive_test = test.copy()
```

```python
naive_test['naive'] = np.asarray(train['Price'])[len(np.asarray(train['Price']))-1]
naive_test['naive'].head()
```

```python
plt.figure(figsize=(14,6))
plt.plot(naive_train['Price'], label='Train')
plt.plot(test['Price'], label='Test')
plt.plot(naive_test['naive'], label='Naive Forecast on Test data')
plt.legend(loc='best')
plt.title('Naive Forecast')
plt.grid()
plt.show()
```

```python
mape_model2_test = mape(test['Price'].values, naive_test['naive'].values)
print('For naive Forecast on Test Data, MAPE is %3.3f'%(mape_model2_test),'%')
```

```python
resultsDf_2 = pd.DataFrame({'Test Mape (%)' : [mape_model2_test]}, index=['NaiveModel'])
result = pd.concat([result, resultsDf_2])
result
```

```python
# now we have got both the value for both the model Linear reg and Naive Bayes
# Now we will create our final model i.e we will preform final forecasting
final_model = ExponentialSmoothing(df,trend='add',seasonal='additive')
final_model_fit = final_model.fit(smoothing_level=0.4, smoothing_trend=0.3, smoothing_seasonal=0.6)
```

```python
Mape_final_model = mape(df['Price'].values, final_model_fit.fittedvalues)
print('MAPE: ', Mape_final_model)
```

```python
predictions = final_model_fit.forecast(steps=len(test))
```

```python
pred_df = pd.DataFrame({'lower_CI': predictions - 1.96*np.std(final_model_fit.resid, ddof=1),
                        'prediction':predictions,
                        'upper_CI': predictions + 1.96*np.std(final_model_fit.resid, ddof=1)})
```

```python
pred_df.head()
```

```python
axis = df.plot(label = 'Actual', figsize = (16,6))
pred_df['prediction'].plot(ax=axis, label='Forecast', alpha=0.5)
axis.fill_between(pred_df.index, pred_df['lower_CI'], pred_df['upper_CI'], color='m', alpha=0.15)
axis.set_xlabel('year-month')
axis.set_ylabel('price')
plt.legend(loc='best')
plt.grid()
plt.show()
```
